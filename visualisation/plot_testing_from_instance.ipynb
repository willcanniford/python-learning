```python
import matplotlib.pyplot as plt
```

```python
class Test():
    def __init__(self):
        pass
    
    def plot(self, ax):
        plot = ax.plot(range(1,11), range(1,11))
        return plot
```

https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10,6))
Test().plot(ax[0][1])
fig.subplots_adjust(top=.9, right=.9, left=.1)
fig.suptitle('Title')
plt.show()
```
