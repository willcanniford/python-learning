# Visualisations with `seaborn`
`matplotlib` provides the basic building blocks for `seaborn` and can be tied in with `pandas` as well; this is particularly useful if you are working with `DataFrame` objects having read in the data that you are working with. 

You can see information about installing `seaborn` [here](https://seaborn.pydata.org/installing.html?highlight=install).

Once you've done that then you can find official documentation and starting tutorial [here](https://seaborn.pydata.org/tutorial.html). 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

I am going to read in some data so that we have something to plot with when using `seaborn`. 

```python
spine = pd.read_csv('../data/Dataset_spine.csv').loc[:, :'Col12']
spine.head()
```

## `distplot()` and `hist()`

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 3))

plt.sca(ax1)
spine.Col1.hist()
plt.title('Pandas hist()')
#plt.show()

plt.sca(ax2)
sns.distplot(spine.Col1)
plt.title('Seaborn distplot()')
plt.show()
```

We can see that we have default stying that comes with `seaborn` as well as a kde on top of the regular binned histogram. 

Additionally, we can leverage `plt.subplots()` with `distplot()` to loop through our variables and see the distribution of each variable. This could be ideal when you are looking to see whether any of them need normalising etc pre model building.

```python
fig, axarr = plt.subplots(4,3, figsize=(12, 12))
plt.subplots_adjust(hspace=0.3, wspace=0.3)

for i, ax in enumerate(axarr.flatten()):
    plt.sca(ax)
    sns.distplot(spine.iloc[:, i])
    
plt.show()
```
